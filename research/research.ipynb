```python
## Explanation is for ensure annotation

def get_product(x:int,y:int) -> int:
    return x*y
```

```python
get_product(2,2)
```

```python
get_product(3,"6")
```

```python
from ensure import ensure_annotations

@ensure_annotations
def get_product(x:int,y:int) -> int:
    return x*y
```

```python
get_product(2,"4")
```
